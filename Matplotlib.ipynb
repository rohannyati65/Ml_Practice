## Bar

```python
# You can make plots from a dictionary
nut_butter_prices = {"Almond butter": 10,
                     "Peanut butter": 8,
                     "Cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Dan's Nut Butter Store",xlabel="Different Butter" ,ylabel="Price ($)");
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

## Histogram

```python
# Make some data from a normal distribution
x = np.random.randn(1000) # pulls data from a normal distribution

fig, ax = plt.subplots()
ax.hist(x);
```

```python
x = np.random.random(1000) # random data from random distribution

fig, ax = plt.subplots()
ax.hist(x);
```

## Subplots

Option 1. Create multiple subplots

```python
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,5))

ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

Option 2: Create multiple subplots

```python
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(10,5))

ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

## Plotting Data directly with Pandas

```python
car_sales=pd.read_csv("car-sales.csv")
car_sales.head()
```

```python
car_sales["Price"]=car_sales["Price"].str.replace('[\$\,\.]','')
car_sales
```

```python
car_sales["Price"]=car_sales["Price"].str[:-2]
car_sales
```

```python
car_sales["Sale Date"]=pd.date_range("1/1/2020",periods=len(car_sales))
car_sales
```

```python
car_sales["Total Sales"]=car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
car_sales["Price"]=car_sales["Price"].astype(int)
car_sales.plot(x="Odometer (KM)",y="Price",kind="scatter");
```

```python
heart_disease=pd.read_csv('heart-disease.csv')
heart_disease.head()
```

```python
heart_disease["age"].plot.hist(bins=20);
```

```python
heart_disease.plot.hist(figsize=(10,30),subplots=True);
```

```python
over_50=heart_disease[heart_disease["age"]>50]
over_50
```

```python
over_50.plot(kind="scatter",x='age',y='chol',c='target',figsize=(10,6));
```

```python
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',x='age',y='chol',c='target',ax=ax);
ax.set_xlim([45,100]);
```

```python
fig,ax=plt.subplots(figsize=(10,6))
scatter=ax.scatter(over_50["age"],over_50["chol"],c=over_50["target"])
ax.set(title="HD AND CL",xlabel="Age",ylabel="Cholesterol");
ax.legend(*scatter.legend_elements(),title="Target");
```

```python
fig,ax=plt.subplots(figsize=(10,6))
scatter=ax.scatter(over_50["age"],over_50["chol"],c=over_50["target"])
ax.set(title="HD AND CL",xlabel="Age",ylabel="Cholesterol");
ax.legend(*scatter.legend_elements(),title="Target");
ax.axhline(over_50["chol"].mean(),color="red",linestyle="--");
```
